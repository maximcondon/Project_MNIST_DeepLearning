# Image Recognition

### VGG16

```python
from keras.preprocessing.image import load_img, img_to_array
from keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
```

```python
m = VGG16()
```

```python
image = load_img('Dolphin.jpg', target_size=(224,224))
```

```python
image
```

```python
image = img_to_array(image)
```

```python
image.shape
```

```python
image = image.reshape((1,224,224,3))
```

```python
image = preprocess_input(image)
```

```python
yhat = m.predict(image)
```

```python
best_prediction = decode_predictions(yhat)
```

```python
print(f'Our best guess is: {best_prediction[0]}')
```
