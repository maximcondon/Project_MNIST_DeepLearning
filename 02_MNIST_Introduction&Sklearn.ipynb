### Let's use an old familiar technique for classifying each images: RandomForestClassifier!

```python
#import the model
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
#instantiate the model
m = RandomForestClassifier(n_estimators=10)
```

```python
#fit the model
m.fit(xtrain, ytrain)
```

```python
m.score(xtrain,ytrain)
#training score -- overfit????
```

```python
m.score(xtest.reshape(-1,784), ytest)
```

```python
ypred = m.predict(xtest.reshape(-1,784))
```

```python
ypred
```

```python
ytest
```

```python
accuracy_score(ytest,ypred)
```

### Gradient-Boosted Random Forest

```python
#import the model
from sklearn.ensemble import GradientBoostingClassifier
```

```python
#instantiate the model
#verbosity will print out information while the model trains
gbc = GradientBoostingClassifier(verbose=1)
```

```python
#fit the model
#only use 5000 data points to speed up the process
#really slow!!!
gbc.fit(xtrain,ytrain)
```

```python
gbc.score(xtrain, ytrain)
#training score
```

```python
gbc.score(xtest.reshape(-1, 784), ytest)
#not bad for only having trained on 5000 data points!
```

### Let's optimize our RandomForest Hyperparameters using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {   
            'n_estimators' : [5, 10, 15, 20, 50, 100],
            'max_depth' : [3, 5, 10]  
}

grid_rf = GridSearchCV(m, param_grid, verbose=1)
```

```python
grid_rf.fit(xtrain, ytrain)
```

```python
grid_rf_best = grid_rf.best_estimator_
```

```python
grid_rf.best_params_
```

```python
grid_rf_best
```

```python
grid_rf_best.score(xtest.reshape(-1, 784), ytest)
```
