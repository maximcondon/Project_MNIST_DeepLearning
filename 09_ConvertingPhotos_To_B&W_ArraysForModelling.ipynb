# Converting images to arrays for modelling:

```python
# Python Imaging Library
from PIL import Image
import numpy as np
```

```python
pwd
```

```python
im = Image.open('/Users/maximcondon/Desktop/Spiced/09_Week_9/hokusaifuji.jpg')
im.size
```

## Can change the size of the photo:

```python
small = im.resize((600,406))
small
```

## Can even crop!
- Left needs to be smaller than right and top needs to be smaller than bottom!

```python
fuji = small.crop((300, 150, 600, 400)) # ((left, top, right, bottom))
                                       # left < right; top < bottom
fuji
```

## Can convert photo to Black and White!

```python
fuji_bw = fuji.convert('L') # black-white image
fuji_bw
```

```python
a = np.array(fuji)
```

```python
a.shape
```

```python
bw = np.array(fuji_bw)
bw.shape
```
